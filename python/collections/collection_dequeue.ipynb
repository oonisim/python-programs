# [collections.deque](https://docs.python.org/2/library/collections.html#collections.deque)

> Returns a new deque object initialized **left-to-right** (using append()) with data from iterable.
> * append(x)  
Add x to the right side of the deque.
> * pop()  
Remove and return an element from the right side of the deque. If no elements are present, raises an IndexError.
> * extend(iterable)  
Extend the right side of the deque by appending elements from the iterable argument.



* [collections.deque()](https://www.hackerrank.com/challenges/py-collections-deque/problem)

> A deque is a double-ended queue. It can be used to add or remove elements from both ends.  
> ```input/left <------> output/right```

```
from collections import deque
d = deque()

# Queue <- 1
d.append(1)
print d
---
deque([1])

# 2 -> queue
d.appendleft(2)
print d
---
deque([2, 1])

d.clear()
print d
---
deque([])

# Queue <-- 1
d.extend('1')
print d
---
deque(['1'])

# 4 3 2 --> Queue[1]
d.extendleft('234')
print d
---
deque(['4', '3', '2', '1'])
d.count('1')
---
1

# Queue[4,3,2] -> 1
d.pop()
---
'1'

print d
---
deque(['4', '3', '2'])

# 4 <- Queue[3, 2]
d.popleft()
---
'4'

print d
---
deque(['3', '2'])

d.extend('7896')
print d
---
deque(['3', '2', '7', '8', '9', '6'])

d.remove('2')
print d
---
deque(['3', '7', '8', '9', '6'])

d.reverse()
print d
---
deque(['6', '9', '8', '7', '3'])

d.rotate(3)
print d
---
deque(['8', '7', '3', '6', '9'])
```

# Example

Perform append, pop, popleft and appendleft methods on an empty deque .

## Input Format

The first line contains an integer , the number of operations.
The next  lines contains the space separated names of methods and their values.

### Sample Input

6
append 1
append 2
append 3
appendleft 4
pop
popleft

## Output Format

Print the space separated elements of deque .

### Sample Output

1 2

````python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import sys
from collections import deque


num_lines = int(input())

queue = deque()

def _append(values):
    queue.append(values)
    
def _pop(values=None):
    queue.pop()

def _popleft(values=None):
    queue.popleft()
    
def _appendleft(values):
    queue.appendleft(values)
    
def _na(values=None):
    pass

commands = {
    "append": _append,
    "pop": _pop,
    "popleft": _popleft,
    "appendleft": _appendleft,
}
    
for line in sys.stdin:
    line = line.rstrip() + " _DUMMY"  # intentionally append dummy   
    words = line.split()
    commands.get(words[0], _na)(words[1])

print(*(list(queue)))
````
