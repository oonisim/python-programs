# TypedDict Class

```python
from typing import (
    TypedDict, Literal, Annotated, List
)
from operator import (
    add
)
```

```python
class State(TypedDict):
    """Chat State (message)
    {
        "messages": List[str]
    }
    """
    # Annotated is to add metadata to the data type, which is List here.
    # Annotated[T, metadata, metadata, ...] annotates the date type T with multiple metadata.
    # Metadata is for the LangGraph to inspect/use. Remember, the data itself is List.
    messages: Annotated[    
        List[str],      # State data
        add    # Reducer (metadata)
    ]
```

```python
state = State(
    message=["Initial"]    # Same as initializing a class with initialil arguments for def __init__(message:Liste[str)
)
```
