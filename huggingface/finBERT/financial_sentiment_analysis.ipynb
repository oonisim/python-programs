# FinBERT - Financial Sentiment Prediction

* [Github FinBERT: Financial Sentiment Analysis with BERT](https://github.com/ProsusAI/finBERT)

> FinBERT is a pre-trained NLP model to analyze sentiment of financial text. ### Reuters Corpora (TRC2)

* [Reuters Corpora (RCV1, RCV2, TRC2)](https://trec.nist.gov/data/reuters/reuters.html)

> In 2000, Reuters Ltd made available a large collection of Reuters News stories for use in research and development of natural language processing, information retrieval, and machine learning systems. This corpus, known as "Reuters Corpus, Volume 1" or RCV1, is significantly larger than the older, well-known Reuters-21578 collection heavily used in the text classification community.

> TRC2: Thomson Reuters Text Research Collection (TRC2)<br>
><br>
>The TRC2 corpus comprises 1,800,370 news stories covering the period from 2008-01-01 00:00:03 to 2009-02-28 23:54:14 or 2,871,075,221 bytes, and was initially made available to participants of the 2009 blog track at the Text Retrieval Conference (TREC), to supplement the BLOGS08 corpus (that contains results of a large blog crawl carried out at the University of Glasgow). TRC2 is distributed via web download.

````python
import os
import logging
import numpy as np
import torch
from transformers import (
    AutoModelForSequenceClassification,
    BertTokenizer,
    pipeline
)
from finbert.finbert import predict
import nltk

logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)
nltk.download('punkt')
````

````python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x - np.max(x, axis=1)[:, None])
    return e_x / np.sum(e_x, axis=1)[:, None]
````

# FinBERT trained model

FinBERT have been trined on the financial news based on the [AutoModelForSequenceClassification](https://huggingface.co/transformers/model_doc/auto.html?highlight=automodelforsequenceclassification#transformers.AutoModelForSequenceClassification) model for the sequence classification tasks. The model takes ```tokens``` and returns [SequenceClassifierOutput](https://huggingface.co/transformers/main_classes/output.html?highlight=sequenceclassifieroutput#seq2seqsequenceclassifieroutput).

````python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
````

## Summary of the news

Generate an overview of the news using ML/NLP.

* NLP = Natural Language Processing

````python
summarizer = pipeline("summarization")
````

````python
summary = summarizer(text, max_length=100, min_length=30, do_sample=False, num_beams=3)
````

````python
for sentence in summary[0]["summary_text"].split('. '):
    print(f"{sentence.strip()}")
````

## Sentiment

Analyze the text to decide the sentiment.

````python
sentiments = {0: 'Positive', 1: 'Negative', 2: 'Neutral'}
````

````python
tokens = tokenizer(
    text,
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="pt"
)
````

````python
with torch.no_grad():
    logits = model(**tokens).logits
````

````python
scores = softmax(np.array(logits))
prediction = sentiments[np.argmax(scores)]
print(f"Sentiment:{prediction}")
print(f"Positive:{scores[0][0]:.02f} Negative:{scores[0][1]:.02f} Neutral:{scores[0][2]:.02f}")
````
