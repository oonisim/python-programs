# Normal distribution

```python
import scipy.stats


u = 2.97
sd = 1
p = scipy.stats.norm(loc=u, scale=sd)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

x = np.linspace(u - 3*sd, u + 3*sd, 100)
plt.plot(x, p.pdf(x))
plt.grid()
plt.show()
```
