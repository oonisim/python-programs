# Meaning of $f(n-t)$

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
fs = 4000
xmin = -40
xmax = 40
n = 15
t = np.arange(xmin,xmax,1/fs)
````

# f(t)

````python
def f(x):
    return (x+10) * x * (x-30)
````

````python
plt.plot(t, f(t))
plt.grid()
````

# f(n-t) is Time Reverse at t = $\frac {n}{2}$

<img src="flip_at_n_2.jpeg" align="left" width=400/>

````python
plt.plot(t, f(t), linestyle='-', color='k', label='f(t)')
plt.plot(t, f(n-t), linestyle='-', color='g', label='f(n-t)')
plt.axvline(x=n/2, color='g', linestyle='--', alpha=0.4, label="t=n/2")

plt.legend()
plt.grid()
plt.title("Flip f(t) at t = n/2")
````

# f(n-t) is Shift n on Time Reverse at t=0

1. Flip $f(t)$ at ```t=0``` $\rightarrow f(-t)$
2. Shift ```+n``` on $f(-t)$ $\rightarrow f(n-t)$

````python
plt.plot(t, f(t), linestyle='--', color='k', label='f(t)', alpha=0.2)
plt.plot(t, f(-t), linestyle='-', color='r', label='f(-t)')
plt.plot(t, f(n-t), color='g', label='f(n-t)')
plt.axvline(x=0, color='r', linestyle='--', alpha=0.4, label='t=0')
plt.axvline(x=n, color='r', linestyle='--', alpha=0.4, label='t=n')
plt.arrow(0, 50000, n, 0, head_width=10000, head_length=2, linestyle='--', linewidth=2, color='r', length_includes_head=True)
plt.legend()
plt.grid()
plt.title("Shift f(-t) to right for n units where (n-t) = 0")
````
