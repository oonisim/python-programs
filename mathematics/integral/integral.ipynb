# Integral

* [scipy.integrate](https://docs.scipy.org/doc/scipy/tutorial/integrate.html)

```python
from typing import (
    Callable
)
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

# Constant

```python
sampling_rate: float = 8000.0
duration: float = np.pi * 2
```

# Function

```python
def f(x):
    return np.sin(x)
```

```python
def generate_samples(sampling_rate: float, seconds: float, func: Callable):
    x = np.linspace(0, seconds, int(sampling_rate * seconds), endpoint=False)
    return x, func(x)
```

```python
x, y = get_samples(sampling_rate=sampling_rate, seconds=duration, func=np.sin)
```

```python
plt.plot(x, y)
plt.xlabel('xin(x)')
plt.ylabel('x')
plt.grid()
plt.show()
```

# $\int_{0}^{\pi} sin(x)$

```python
print(quad(np.sin, 0, np.pi))
```
