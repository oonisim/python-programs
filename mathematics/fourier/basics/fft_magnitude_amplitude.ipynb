```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams.update({'font.size': 10})
```

# Sampling

```python
# Sampling
dt = 0.001
Fs = 1.0 / dt
```

# Signal with two frequencies


```python
t = np.arange(0,1,dt)
f_clean = 2 * np.sin(2*np.pi*50*t) + 0.6 * np.sin(2*np.pi*120*t) # Sum of 2 frequencies 50 Hz and 120 Hz
f_noise = 2.5*np.random.randn(len(t))                  # Noise
f = f_clean + f_noise
```

# FFT

```python
# Fourier coefficient
N = len(f)
fhat = np.fft.rfft(a=f, n=N, norm="backward")
frequencies = np.fft.rfftfreq(n=N, d=dt)
```

# Amplitude

<img src="../image/fft_to_amplitude.png" align="left" width=500/>

<img src="../image/fft_to_power.png" align="left" width=300/>

```python
amplitude = 2 * np.abs(fhat) / N

plt.grid(visible=True, which='major', color='k', linestyle='-')
plt.grid(visible=True, which='minor', color='k', linestyle='--')
plt.minorticks_on()
plt.plot(frequencies, amplitude)
plt.show()
```

```python
A = 2 * np.sqrt(fhat * np.conj(fhat)) / N
print(A[1])

plt.grid(visible=True, which='major', color='k', linestyle='-')
plt.grid(visible=True, which='minor', color='k', linestyle='--')
plt.minorticks_on()
plt.plot(frequencies, A)
plt.show()
```
