```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 50

# Parameters for the decaying impulse response
n = np.arange(0, 1, 1/Fs)  # Time samples
decay_rate = 10  # Decay rate
impulse_response = np.exp(-decay_rate * n)  # Exponentially decaying response

# Plot the impulse response
plt.stem((n*Fs), impulse_response)
plt.title('Decaying Impulse Response')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
y = 2 * np.fft.rfft(impulse_response, norm="forward")
freq = np.fft.rfftfreq(len(impulse_response), d=1/Fs)
```

```python
plt.stem(freq, np.abs(y))
plt.title('FFT on Decaying Impulse Response')
plt.xlabel('Freq')
plt.ylabel('Amplitude')
plt.grid()
```
