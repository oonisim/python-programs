```python
def func(x):
    return 0.5 * np.exp(x) + 1
a, b = 0.5, 1.5
x = np.linspace(0, 2)
y = func(x)
Ix = np.linspace(a, b)
Iy = func(Ix) # <6>
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$',
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
# plt.savefig('../../images/ch07/mpl_19')
```

## Static 3D Plotting

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
```

```python
strike, ttm = np.meshgrid(strike, ttm)
```

```python
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}, figsize=(8, 8))
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.coolwarm, linewidth=0.5,
                       antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5);
# plt.savefig('../../images/ch07/mpl_20')
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir='z', s=25,
           c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');
# plt.savefig('../../images/ch07/mpl_21')
```

## Interactive 2D Plotting

### Basic Plots

```python
import pandas as pd
```

```python
import cufflinks as cf
```

```python
cf.set_config_file(offline=True)
```

```python
a = np.random.standard_normal((250, 5)).cumsum(axis=0)
```

```python
index = pd.date_range('2019-1-1',
                      freq='B',
                      periods=len(a)) # <4> 
```

```python
df = pd.DataFrame(100 + 5 * a,
                  columns=list('abcde'),
                  index=index)
```

```python
df.head()
```

```python
df.plot()
```

```python
df.iplot()
```

```python
df[['a', 'b']].iplot(
             theme='polar',
             title='A Time Series Plot',
             xTitle='date',
             yTitle='value',
             mode={'a': 'markers', 'b': 'lines+markers'},
             symbol={'a': 'cross', 'b': 'diamond'},
             size=3.5,
             colors={'a': 'blue', 'b': 'magenta'},
                        )
```

```python
df.iplot(kind='hist',
             subplots=True,
             bins=15,
        )
```

### Financial Plotting

```python
# data from FXCM Forex Capital Markets Ltd.
raw = pd.read_csv('http://hilpisch.com/fxcm_eur_usd_eod_data.csv',
                 index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
quotes = raw[['AskOpen', 'AskHigh', 'AskLow', 'AskClose']]
quotes = quotes.iloc[-60:]
quotes.tail()
```

```python
qf = cf.QuantFig(
         quotes,
         title='EUR/USD Exchange Rate',
         legend='top',
         name='EUR/USD'
)
```

```python
qf.iplot()
```

```python
qf.add_bollinger_bands(periods=15,
                       boll_std=2)
```

```python
qf.iplot()
```

```python
qf.add_rsi(periods=14,
          showbands=False)
```

```python
qf.iplot()
```
