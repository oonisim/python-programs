# Insert column at specific position

1. [pandas.DataFrame.insert](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html)
2. Reindex 

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json(
    "./data/recovery.json",
    dtype={
        "facility": pd.CategoricalDtype(), # Must specify a class instance (https://stackoverflow.com/questions/75089569)
        "supplier": pd.CategoricalDtype(),
        "supplierCode": pd.CategoricalDtype(),
        "suppliedM3": np.float32,
        "recoveredM3": np.float32,
    },
    convert_dates=['date']
)
df
```

```python
df.info()
```

## Insert at position

```python
df.insert(
    loc=3,
    column='elapsed',
    value=pd.to_timedelta('00:' + df['processTime'], errors='coerce'),
    allow_duplicates=False
)
df
```
