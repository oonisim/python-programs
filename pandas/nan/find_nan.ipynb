# Method

[pandas.isnull](https://pandas.pydata.org/docs/reference/api/pandas.isnull.html) (```isna``` is an alias of ```isnull```)

> Returns: bool or **array-like** of bool  
> Return scalar boolean for scalar input, or **an array of boolean** indicating whether each corresponding element is missing.


* object.```isnull()``` method returns boolean array in the same shape of the object. 
* Use [.any(axis=0|1)](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.any.html) or [.all(axis=0|1)](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.all.html) methods on the array returned by ```.isnull()``` for row or column wise handling.



## Dataframe

Return the bollean datafame indicating each cell is NaN/NaT or not.

````python
df.isnull().shape
````

## Row
Use ```(axix=1)``` to find if a row has NA cell or not.

````python
df[df.isnull().any(axis=1)]  # Find a row that has NA cell
````

````python
df[df.isnull().sum(axis=1) > 0]  # Show rows that has at least 1 NA cell
````

````python
df[df.isnull().any(axis=1)].isnull().sum(axis=1)  # Number of NA cells per each row that has NA
````

````python
df.isnull().all(axis=1)  # Find a row where all cells are NA
````

````python
df.iloc[0, :].isnull().any()  # If 0-th row has NA cells
````

````python
list(df.isnull().any(axis=1).index)
````

## Column
Use ```(axix=0)``` to find if a column has NA cell or not.

````python
df.isnull().any(axis=0)   # Which columns have NA cell
````

````python
df.isnull().sum(axis=0)    # How many NA cells each column has
````

````python
df.iloc[:, 0].isnull().any()  # If 0-th column has NA cells
````
