# Display Dataframe as Table

* [Show DataFrame as table in iPython Notebook](https://stackoverflow.com/a/29665452/4281353)

```python
import numpy as np
import pandas as pd
from IPython.display import (
    display,
    HTML,
)
```

```python
df = pd.read_json(
    "./data/recovery.json",
    dtype={
        "facility": pd.CategoricalDtype(), # Must specify a class instance (https://stackoverflow.com/questions/75089569)
        "supplier": pd.CategoricalDtype(),
        "supplierCode": pd.CategoricalDtype(),
        "suppliedM3": np.float32,
        "recoveredM3": np.float32,
    },
    convert_dates=['date']
)
```

```python
DEBUG: bool = True
```

```python
if DEBUG:
    display(df)
```

```python
if DEBUG:
    display(HTML(df.to_html()))
```
