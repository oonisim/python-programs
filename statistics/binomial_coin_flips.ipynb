```python
import math
from scipy.stats import binom
import matplotlib.pyplot as p
import numpy as np
```

```python
prob = 0.5    # coin head probability
size = 100    # number of flips
```

```python
# Plot for 100 coin flips
x = np.arange(0, size+1, 1)
y = binom.pmf(x, size+1, prob)
```

```python
fig, ax1 = p.subplots(figsize=(5, 3))

ax1.set_xticks([0, 20, 40, 60, 80, 100])
ax1.set_ylabel("P(x) for 100 Flips")
ax1.set_yticks([0,0.02,0.04,0.06,0.08])
#ax1.set_yticks([0,0.05,0.1,0.15,0.2,0.25])
ax1.plot(x,y, color="tab:blue")

x_lt_40 = np.arange(0, 41, 1)
x_gt_60 = np.arange(61, size+1, 1)
p.fill_between(x_lt_40, y[:41], facecolor='blue', alpha=0.5)
p.fill_between(x_gt_60, y[61:], facecolor='blue', alpha=0.5)

fig.tight_layout()
p.grid()
p.show()
```

# p-value of 61 heads out of 100

```python
def cumulative_probability_of_r_out_of_n(n, lower, upper) -> float:
    """Cumulative probability of r between lower and upper.
    e.g. sum of the probabilities of heads between 40 and 60 out of 100 flips,
    including heads=40 and heads=60.
    """
    return sum([
        math.comb(n, i) / (2 ** n) for i in range(lower, upper+1)
    ])
```

```python
1 - cumulative_probability_of_r_out_of_n(n=100, lower=40, upper=60)
```

```python
1 - cumulative_probability_of_r_out_of_n(n=100, lower=41, upper=59)
```

# 100 flips vs 1000 flips

```python
prob = 0.5    # coin head probability
```

```python
# Plot for 100 coin flips
size_1 = 100    # number of flips
x_1 = np.arange(0, size_1+1, 1)
y_1 = binom.pmf(x_1, size_1+1, prob)
```

```python
# Plot for 1000 coin flips
size_2 = 1000    # number of flips
x_2 = np.arange(0, size_2+1, 1)
y_2 = binom.pmf(x_2, size_2+1, prob)
```

```python
fig, ax1 = p.subplots(figsize=(10, 5))

ax1.set_xticks([0, 200, 400, 600, 800, 1000])
ax1.set_xlabel("0 - 1000 Flips", color="tab:red")
ax1.set_yticks([0,0.02,0.04,0.06,0.08])
ax1.set_ylabel("P(x) for n Flips")
ax1.tick_params(axis="x", colors="tab:red")
x_lt_469 = np.arange(0, 470, 1)
x_gt_531 = np.arange(532, size_2+1, 1)
p.fill_between(x_lt_469, y_2[:470], facecolor='red', alpha=0.5)
p.fill_between(x_gt_531, y_2[532:], facecolor='red', alpha=0.5)

ax2 = ax1.twiny()
ax2.set_xticks([0, 20, 40, 60, 80, 100])
ax2.set_xlabel("0 - 100 Flips", color="tab:blue")
# ax2.set_yticks([0,0.05,0.1,0.15,0.2,0.25])
ax2.set_yticks([0,0.02,0.04,0.06,0.08])
ax2.tick_params(axis="x", colors="tab:blue")
x_lt_40 = np.arange(0, 41, 1)
x_gt_60 = np.arange(61, size+1, 1)
p.fill_between(x_lt_40, y_1[:41], facecolor='blue', alpha=0.5)
p.fill_between(x_gt_60, y_1[61:], facecolor='blue', alpha=0.5)

ax2.plot(x_1,y_1, color="tab:blue")
ax2.grid()
ax1.plot(x_2,y_2, color="tab:red")
ax1.grid()

fig.tight_layout()
p.grid()
p.show()
```
