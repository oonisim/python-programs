```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
MEAN = 3.0
SCALE = 1.0
```

```python
x = np.linspace(MEAN - 2.0, MEAN+2.0, 100)
fig, ax = plt.subplots(1, 1)
ax.plot(
    x, 
    norm.pdf(x, loc=MEAN, scale=SCALE),
    'r-', lw=5, alpha=0.6, label='norm pdf p(θ)'
)
ax.vlines(4.0, 0.02, 0.4, linestyle='dashed')
ax.legend()
ax.grid()
```

# Likelihood

Likelihood $p( \theta = 4.0) $ = 0.242.

```python
norm.pdf(4.0, loc=MEAN, scale=SCALE)
```
