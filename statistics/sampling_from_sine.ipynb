```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 1
bins = 300
n = 100000000
frequency = 1
```

```python
theta = np.linspace(0, 2 * np.pi, n)
```

```python
x = np.sin(frequency * theta)
weights = np.ones_like(x) / len(x)
```

```python
plt.plot(theta / np.pi, x)
plt.xlabel('pi')
plt.ylabel(f'sin({frequency}Î¸)')
plt.grid()
```

```python
plt.xlabel('x')
plt.ylabel(f'p(x)')
plt.grid()
_ = plt.hist(x, bins=bins, weights=weights, color='b')
```
