# RVS vs PDF

RVS = Randome Variable Sampling = Get samples $x_i$ from a distribution $X$.

PDF($x_i$)= Probability to get $x_i$ from $X$.

<img src="./image/rvs_vs_pdf.png" align="left"/>

```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Random Variable from Uniform Function

```python
u = stats.uniform()
plt.plot(range(20), u.rvs(size=20))
```

# Probability of random variable x from Uniform Function

```python
plt.plot(range(20), u.pdf(u.rvs(size=20)))
```
