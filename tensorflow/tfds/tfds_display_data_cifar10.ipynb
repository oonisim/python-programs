# Display TFDS data examples

* [tfds.visualization.show_examples](https://www.tensorflow.org/datasets/api_docs/python/tfds/visualization/show_examples)

> Visualize images (and labels) from an image classification dataset.

```
tfds.visualization.show_examples(
    ds: tf.data.Dataset, ds_info: dataset_info.DatasetInfo, **options_kwargs
)
```



````python
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_addons as tfa
````

````python
ds, ds_info = tfds.load('cifar10', split='train', with_info=True)
fig = tfds.show_examples(ds, ds_info)
````
