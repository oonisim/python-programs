```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = 4000
var_x = 20
sigma_x = np.sqrt(var_x)
```

```python
v = 280
var_v = 5
sigma_v = np.sqrt(var_v)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3)) # 1 row 2 columns

_ = np.linspace(x - 3*sigma_x, x + 3*sigma_x, 100)
axes[0].title.set_text('σx')
axes[0].plot(_, stats.norm.pdf(_, x, sigma_x))
axes[0].grid()

_ = np.linspace(v - 3*sigma_v, v + 3*sigma_v, 100)
axes[1].title.set_text('σv')
axes[1].plot(_, stats.norm.pdf(_, v, sigma_v))
axes[1].grid()
```
