# Display images

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,8)) 

# --------------------------------------------------------------------------------
# arctan2(y/x) with y=1 fixed on a scauel with unit=)
# --------------------------------------------------------------------------------
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].spines['bottom'].set_position('zero')
ax[0].spines['left'].set_position('zero')
ax[0].set_xlabel("x", loc='right', fontsize=13)
ax[0].set_ylabel("", loc="top", fontsize=18)
ax[0].set_title("arctan2(y/x) with fixed y=1  (unit is )", fontsize=15)
ax[0].tick_params(axis='x', colors='red')
ax[0].grid()

x = np.linspace(-10, 10, 10001)[::-1]
y1 = 1
ax[0].plot(x, np.arctan2(y1, x) / np.pi, label="atan2(y=+1/x)")
y2 = -1
ax[0].plot(x, np.arctan2(y2, x) / np.pi, label="atan2(y=-1/x)")
ax[0].legend()

# --------------------------------------------------------------------------------
#  = arctan2(y=1/x) increase as x moves from + to -.
# --------------------------------------------------------------------------------
ax[1].spines['top'].set_visible(False)
ax[1].spines['bottom'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].tick_params(
    axis='both',
    which='both',
    bottom=False,
    top=False,
    left=False, 
    right=False,
    labelbottom=False,
    labelleft=False
) 
img = mpimg.imread('image/atan2_as_x_move_with_fixed_y.png')
imgplot = ax[1].imshow(img)
```
