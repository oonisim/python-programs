```python
import matplotlib.pyplot as p
import numpy as n
from scipy.special import comb as c
 
fig, ax1 = p.subplots()
 
# Plot for 10 coin flips
x = n.linspace(0, 100, 1000)
y = c(10, x/10)*0.5**(x/10)*0.5**(10-x/10)
ax1.set_xlabel("% Heads for 10 flips", color="tab:blue")
ax1.set_xticks([0,25,50,75,100])
ax1.set_ylabel("P(x) for 10 Flips", color="tab:blue")
ax1.set_yticks([0,0.05,0.1,0.15,0.2,0.25])
ax1.tick_params(axis="x", colors="tab:blue")
ax1.tick_params(axis="y", colors="tab:blue")
ax1.plot(x,y, color="tab:blue")
 
# Plot for 100 coin flips
x = n.linspace(0, 100, 1000)
y = c(100, x)*0.5**(x)*0.5**(100-x)
ax2 = ax1.twinx()
ax2.set_xticks([0,25,50,75,100])
ax2.set_xlabel("% Heads for 100 flips", color="tab:red")
ax2.set_ylabel("P(x) for 100 Flips", color="tab:red")
ax2.set_yticks([0,0.02,0.04,0.06,0.08])
ax2.tick_params(axis="x", colors="tab:red")
ax2.tick_params(axis="y", colors="tab:red")
ax2.plot(x,y, color="tab:red")
 
fig.tight_layout()
p.show()
```
